```python
#Flask-1 Assignment
"""Q1. What is Flask Framework? What are the advantages of Flask Framework?"""
Ans: Flask is a micro web framework written in Python. It is designed to make getting started quick and easy, with the 
ability to scale up to complex applications. Flask provides tools, libraries, and technologies that allow developers to
build a web application.

Advantages of Flask Framework:

1. Easy to get started: Flask is easy to set up and get started with. It has a simple and minimalistic design, which 
makes it easy to learn and use.

2. Flexible: Flask is highly flexible and allows developers to customize the framework according to their needs.

3. Lightweight: Flask is a lightweight framework, which makes it ideal for developing small applications.

4. Scalable: Flask is highly scalable and can be used to develop large applications.

5. Open source: Flask is an open source framework, which means that developers can access the source code and modify it
according to their needs.

6. Extensible: Flask is highly extensible and allows developers to add additional features and functionality to the 
framework.

"""Q2. Create a simple Flask application to display ‘Hello World!!’. Attach the screenshot of the output in
Jupyter Notebook."""
Ans: 
from flask import Flask

app = Flask(__name__)

@app.route("/")
def hello_world():
    return "<h1>Hello, World!</h1>"

if __name__=="__main__":
    app.run(host="0.0.0.0")
    
"""Q3. What is App routing in Flask? Why do we use app routes?"""
Ans: App routing in Flask is the process of mapping URLs to functions that are defined in the application. It is used to
create a web application by defining the URLs that should be associated with each function. App routing allows us to 
create different pages for our web application, and it also allows us to create dynamic URLs that can be used to access 
specific resources. App routing is also used to create custom error pages, which can be used to display helpful messages
when an error occurs. App routing is essential for creating a web application as it allows us to control how users 
interact with our application.

"""Q4. Create a “/welcome” route to display the welcome message “Welcome to ABC Corporation” and a “/”
route to show the following details:
Company Name: ABC Corporation
Location: India
Contact Detail: 999-999-9999"""
Ans: 
from flask import Flask

app = Flask (__name__)

@app.route("/")
def data():
    return "<h3>Company Name: ABC Corporation <br> Location: India <br> Contact Detail: 999-999-9999</h3>"

@app.route("/welcome")
def welcome():
   return "<h1>Welcome to ABC Corporation</h2>"

if __name__=="__main__":
    app.run(host="0.0.0.0")

"""Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the
url_for() function."""
Ans: 
The url_for() function is used in Flask for URL Building.

Example:

from flask import Flask, url_for
app = Flask(__name__)
 
@app.route('/')
def index():
   return 'Index Page'

@app.route('/hello')
def hello():
   return 'Hello World'

@app.route('/login')
def login():
   return 'Login Page'
 
with app.test_request_context():
   print(url_for('index'))
   print(url_for('hello'))
   print(url_for('login'))
if __name__=="__main__":
    app.run(host="0.0.0.0")
    
#Output:
/
/hello
/login
```
